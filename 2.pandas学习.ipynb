## 2.3 Series案例
### 案例1


```python
import numpy as np
import pandas as pd # type: ignore

np.random.seed(2)
values = np.random.randint(50, 101, 10)
indexes = []
for i in range(1, 11):
    indexes.append("学生" + str(i))
# print(indexes)

scores = pd.Series(values, indexes)
# print(scores)
print(f"平均分是：{scores.mean()}分")
print(f"最高分是：{scores.max()}分")
print(f"最低分是：{scores.min()}分")

mean = scores.mean()
upper_scores = scores[scores > mean]        # 布尔索引

# print(upper_scores)
print(f"高于平均分的人数是：{upper_scores.count()}人")
print(f"高于平均分的人数是：{len(upper_scores)}人")
```

```python
import numpy as np
import pandas as pd
temperatures = pd.Series([28, 31, 29, 32, 30, 27, 33], ['周一', '周二', '周三', '周四', '周五', '周六', '周日'])
# print(temperatures)

upper_temperatures = temperatures[temperatures > 30]
print(f"温度超过30℃的天数是：{upper_temperatures.count()}天")
print(f"平均温度是：{temperatures.mean()}℃")

print(temperatures.sort_values(ascending=False))

# output = np.zeros(6)
# for i in range(len(temperatures) + 1):
#     if i == 6:
#         break
#     output[i] = temperatures.iloc[i] - temperatures.iloc[i+1]
# output_abs = np.abs(output)
# print(output_abs)
# print(f"最大温差是：{np.max(output_abs)}℃")
# print(f"最大温差索引是：{np.argmax(output_abs)}")

t3 = temperatures.diff().abs()     # 计算series数组的变化值

t3.sort_values(ascending=False)
print(f"温差变化最大的是：{t3.sort_values(ascending=False).keys()[0]}")


print(t3.index)
t3.values
```

```python
import pandas as pd
# 日期序列
date = pd.date_range("2025-01-1",periods=10)

prices = pd.Series( [102.3, 103.5, 105.1,
                    104.8, 106.2, 107.0, 106.5, 108.1, 109.3, 110.2],
                      index = date)
```

```python
# 收益率
a = prices.pct_change()     # percent
print(a)
# 收益率最高的日期
print(f"收益率最高的日期是：{a.idxmax()}")
# 波动率
print(a.std())
```

```python
import pandas as pd

data = {
    '姓名': ['张三', '李四', '王五', '赵六', '钱七'],
    '数学': [85, 92, 78, 88, 95],
    '英语': [90, 88, 85, 92, 80],
    '物理': [75, 80, 88, 85, 90]
}
scores = pd.DataFrame(data)
scores
```

```python
scores["总分"]  = scores[['数学','英语','物理']].sum(axis=1)
scores["平均分"]  = scores[['数学','英语','物理']].mean(axis=1).round(2)
scores
```

```python
scores[(scores['数学']>90) | (scores['英语']>85)]
```

```python
scores.sort_values('总分',ascending=0).head(3)
```
