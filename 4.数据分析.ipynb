```python
# 缺失值的处理NAN
import  pandas as pd
import  numpy as np
s = pd.Series([12,23,np.nan,None,pd.NA])
df = pd.DataFrame([[1,pd.NA,2], [2,3,5], [None,5,6]],
                  columns=['第一列', '第二列', '第三列'],
                  index=['第一行', '第二行', '第三行'])
print(s)
print(df)
print("-"*30)
# 缺失值的检测
print(s.isna())
print("-"*30)
print(s.isnull())
print("-"*30)
print(df.isna())
print(df.isnull())
```

```python
print(s.isna().sum())
print('-'*30)
print(df.isna().sum())  # 列
print(df.isna().sum(axis=1))  # 行
```

```python
# 剔除缺失值
print(s.dropna())
print("-"*30)
print(df)
print(df.dropna())  # 删掉全部有缺失值的行
print(df.dropna(how='all'))     # 所有的值都是缺失值才删除
print(df.dropna(thresh=2))     # 至少有n个值不是缺失值，才保留
print("-"*30)
print(df.dropna(axis=1))    # 删除有缺失值的列
```

```python
# 数据变形
import pandas as pd

data = {
    'ID': [1, 2],
    'name': ['alice', 'bob'],
    'Math': [90, 85],
    'English': [88, 92],
    'Science': [95, 89]
}
df = pd.DataFrame(data)
df.T
df
```

```python
# 宽表转换成长表
df2 = pd.melt(df,id_vars=['ID','name'],var_name='科目',value_name='分数')
df2.sort_values('name')
```

pd.melt() 的执行过程就像 “拆积木”，步骤如下：
- 保留标识列：先把 id_vars 里的 ID、name 列完整保留，作为每一行的基础标识；
- 拆分非标识列：把剩下的 Math、English、Science 列拆成两部分：
- 列名（Math/English/Science）→ 作为 “科目” 列的取值；
- 列值（90/85/88 等）→ 作为 “分数” 列的取值；
- 重组行：每一个 “标识列组合 + 一个科目 + 一个分数” 形成一行，最终生成长表。

```python
# 长表转宽表
pd.pivot(df2,index=['ID','name'],columns='科目', values='分数').reset_index()
```

```python
# 数据分箱  pd.cut(x,bins,labels)
import pandas as pd
df = pd.read_csv('data/employees.csv')
df.head(10)
```

```python
df1 = df.head(10)[['employee_id', 'salary']]
pd.cut(df1['salary'],bins=2)
# bins为整数：分成n段区间，起始值、结束值是所有数据的最小值、最大值
pd.cut(df1['salary'],bins=3).value_counts() # 统计各个区间的人数
```

```python
# bins = list列表
pd.cut(df1['salary'],bins=[0,10000,20000,30000]).value_counts()
df1['收入范围'] = pd.cut(df1['salary'],bins=[0,10000,20000,30000],labels=['低','中', '高'])
print(df1)

# 等分：每段人数相等
pd.qcut(df1['salary'],3).value_counts()
```
